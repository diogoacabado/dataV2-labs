```python
import numpy as np
import pandas as pd

animals = pd.read_csv('mammals.csv')
animals.head()
```

```python
animals['bodywtkg'] = animals['body'] * 0.45359237
animals.head()
```

```python
animals['wtratio'] = animals['body'] / animals['brain']
animals.head()
```

```python
animals['wtratiozerocheck'] = np.where(animals['brain'] != 0, animals['body'] / animals['brain'], 0)
animals.head()
```

```python
animals['sum'] = animals.sum(axis=1)
animals.head()
```
